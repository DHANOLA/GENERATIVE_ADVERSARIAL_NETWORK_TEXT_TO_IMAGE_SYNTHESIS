# GENERATIVE TEXT TO IMAGE SYNTHESIS PART 3

## IMPORTING THE LIBRARIES

```python
import matplotlib.pyplot as plt
import numpy as np
import os

from PIL import Image as image
from IPython.display import display, Markdown, Latex
```

## IMAGES GENERATED BY GAN DURING THE TRAINING

```python
image_caps_loc='/kaggle/input/generative-text-to-image-synthesis-part-2-epoch-25/\
Data/Flowers/training/GENERATIVE TEXT TO IMAGE SYNTHESIS GAN MODEL/samples/caps.txt'

image_ids_loc='/kaggle/input/generative-text-to-image-synthesis-part-2-epoch-25/\
Data/Flowers/training/GENERATIVE TEXT TO IMAGE SYNTHESIS GAN MODEL/samples/ids.txt'

sample_dir='Data/Flowers/training/GENERATIVE TEXT TO IMAGE SYNTHESIS GAN MODEL/samples/fake_image_'
img_dir='/kaggle/input/generative-text-to-image-synthesis-part-2-epoch-25/\
Data/Flowers/training/GENERATIVE TEXT TO IMAGE SYNTHESIS GAN MODEL/samples/'
```

```python

def example_plot(ax, img, fontsize=12, hide_labels=False):
        pc = ax.imshow(img, interpolation='nearest',aspect='auto')
        return pc

with open(image_caps_loc) as caps_file, open(image_ids_loc) as img_id_file:
    caps_list=caps_file.readlines()
    ids_list=img_id_file.readlines()
    
    
for i in range(len(caps_list)):
    
    fig, axs = plt.subplots(2, 3, layout='constrained', figsize=(10, 6))
    
    gridspec = axs[0, 0].get_subplotspec().get_gridspec()

    # clear the left column for the subfigure:
    for a in axs[:, 0]:
        a.remove()

    # plot data in remaining axes:

    img=np.asarray(image.open('/kaggle/input/generative-text-to-image-synthesis-part-2-epoch-25/'+sample_dir+str(i)+'.jpg'))
    axs[0,1].imshow(img, interpolation='nearest')
    axs[0,1].axis('off')
    axs[0,1].set_title('Upto 25th Epoch')

    img=np.asarray(image.open('/kaggle/input/generative-text-to-image-synthesis-part-2-epoch-50/'+sample_dir+str(i)+'.jpg'))
    axs[0,2].imshow(img,interpolation='nearest')
    axs[0,2].axis('off')
    axs[0,2].set_title('Upto 50th Epoch')
    
    img=np.asarray(image.open('/kaggle/input/generative-text-to-image-synthesis-part-2-epoch-75/'+sample_dir+str(i)+'.jpg'))
    axs[1,1].imshow(img, interpolation='nearest')
    axs[1,1].axis('off')
    axs[1,1].set_title('\nUpto 75th Epoch')

    img=np.asarray(image.open('/kaggle/input/generative-text-to-image-synthesis-part-2-epoch100/'+sample_dir+str(i)+'.jpg'))
    axs[1,2].imshow(img, interpolation='nearest')
    axs[1,2].axis('off')
    axs[1,2].set_title('\nUpto 100th Epoch')

    # make the subfigure in the empty gridspec slots:
    subfig = fig.add_subfigure(gridspec[:, 0])

    axsLeft = subfig.subplots( sharey=True)

    img=np.asarray(image.open(img_dir+str(i)+'_'+ids_list[i].split('\n')[0]+'.jpg'))

    pc = example_plot(axsLeft, img)
    subfig.suptitle('Actual Image', fontsize='xx-large', horizontalalignment='center',verticalalignment='top',x=.5, y=1.1)

    fig.suptitle('Actual V/s Generated Image - '+str(i+1)+'\n', fontsize=25,horizontalalignment='left',verticalalignment='top',x=.2)

    
    fig.set_facecolor('1')
    plt.axis('off')
    plt.show()# __i:__ _klnl_
    display(Markdown('#### __Image Description:__ <u>_'+caps_list[i].split('\n')[0].title()+'_ \n'))
    print('\n'*5)
```

## IMAGES GENERATED BY GAN DURING THE VALIDATION / TESTING AT EACH EPOCH - 25th, 50th, 75th, 100th

```python

def example_plot(ax, img, fontsize=12, hide_labels=False):
        pc = ax.imshow(img, interpolation='nearest',aspect='auto')
        return pc

    
for epoch in range(25,125,25):
    
    if epoch<100:
        epoch_dir='/kaggle/input/generative-text-to-image-synthesis-part-2-epoch-'+ \
        str(epoch)+'/Data/Flowers/training/GENERATIVE TEXT TO IMAGE SYNTHESIS GAN MODEL/val_samples/'
    else:
        epoch_dir='/kaggle/input/generative-text-to-image-synthesis-part-2-epoch'+ \
        str(epoch)+'/Data/Flowers/training/GENERATIVE TEXT TO IMAGE SYNTHESIS GAN MODEL/val_samples/'
    
    for i in range(len(os.listdir(epoch_dir))):
    
    
        fig, axs = plt.subplots(3, 4, layout='constrained', figsize=(15, 8))

        gridspec = axs[0, 0].get_subplotspec().get_gridspec()

        # clear the left column for the subfigure:
        for a in axs[:, 0]:
            a.remove()

        # plot data in remaining axes:

        img=np.asarray(image.open(epoch_dir+os.listdir(epoch_dir)[i]+'/fake_image_0.jpg'))
        axs[0,1].imshow(img, interpolation='nearest')
        axs[0,1].axis('off')
        axs[0,1].set_title('\nImage Generated - 1\n', fontsize='x-large')

        img=np.asarray(image.open(epoch_dir+os.listdir(epoch_dir)[i]+'/fake_image_1.jpg'))
        axs[0,2].imshow(img, interpolation='nearest')
        axs[0,2].axis('off')
        axs[0,2].set_title('\nImage Generated - 2\n', fontsize='x-large')

        img=np.asarray(image.open(epoch_dir+os.listdir(epoch_dir)[i]+'/fake_image_2.jpg'))
        axs[0,3].imshow(img, interpolation='nearest')
        axs[0,3].axis('off')
        axs[0,3].set_title('\nImage Generated - 3\n', fontsize='x-large')

        img=np.asarray(image.open(epoch_dir+os.listdir(epoch_dir)[i]+'/fake_image_3.jpg'))
        axs[1,1].imshow(img, interpolation='nearest')
        axs[1,1].axis('off')
        axs[1,1].set_title('\nImage Generated - 4\n', fontsize='x-large')

        img=np.asarray(image.open(epoch_dir+os.listdir(epoch_dir)[i]+'/fake_image_4.jpg'))
        axs[1,2].imshow(img, interpolation='nearest')
        axs[1,2].axis('off')
        axs[1,2].set_title('\nImage Generated - 5\n', fontsize='x-large')

        img=np.asarray(image.open(epoch_dir+os.listdir(epoch_dir)[i]+'/fake_image_5.jpg'))
        axs[1,3].imshow(img, interpolation='nearest')
        axs[1,3].axis('off')
        axs[1,3].set_title('\nImage Generated - 6\n', fontsize='x-large')

        img=np.asarray(image.open(epoch_dir+os.listdir(epoch_dir)[i]+'/fake_image_6.jpg'))
        axs[2,1].imshow(img, interpolation='nearest')
        axs[2,1].axis('off')
        axs[2,1].set_title('\nImage Generated - 7\n', fontsize='x-large')

        img=np.asarray(image.open(epoch_dir+os.listdir(epoch_dir)[i]+'/fake_image_7.jpg'))
        axs[2,2].imshow(img, interpolation='nearest')
        axs[2,2].axis('off')
        axs[2,2].set_title('\nImage Generated - 8\n', fontsize='x-large')

        img=np.asarray(image.open(epoch_dir+os.listdir(epoch_dir)[i]+'/fake_image_8.jpg'))
        axs[2,3].imshow(img, interpolation='nearest')
        axs[2,3].axis('off')
        axs[2,3].set_title('\nImage Generated - 9\n', fontsize='x-large')


        # make the subfigure in the empty gridspec slots:
        subfig = fig.add_subfigure(gridspec[:, 0])

        axsLeft = subfig.subplots( sharey=True)

        img=np.asarray(image.open(epoch_dir+os.listdir(epoch_dir)[i]+'/'+os.listdir(epoch_dir)[i]+'.jpg'))

        pc = example_plot(axsLeft, img)
        subfig.suptitle('At Epoch '+str(epoch)+' Actual Image - '+str(i+1), fontsize='xx-large', horizontalalignment='center',verticalalignment='top',x=.55, y=1.08)

        fig.suptitle('Vaidation / Testing Image At Epoch - '+str(epoch)+'\n', fontsize=25,horizontalalignment='left',verticalalignment='top',x=.25)

        with open(epoch_dir+os.listdir(epoch_dir)[i]+'/caps.txt') as caps_file, open(image_ids_loc) as img_id_file:
            caps_list=caps_file.readlines()


        fig.set_facecolor('1')
        plt.axis('off')
        plt.show()
        print('\n')
        display(Markdown('#### __Image Description:__ <u>_'+caps_list[0].split('\n')[0].title()+'_'))
        print('\n')
```
